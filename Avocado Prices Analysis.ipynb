### 5th question

```python
cities = []
for i in df['region'].unique():
    cities.append(i)
#     print(i)
```

```python
cities
```

```python
# let's first look at conventional type
filt = df['type']=='conventional'
g = sns.factorplot('AveragePrice','region',data=df[filt],
                   hue='year',
                   size=8,
                   aspect=0.7,
                   palette='Blues',
                   join=False
              )
```

```python
order = (
    df[filt & (df['year']==2018)]
    .groupby('region')['AveragePrice']
    .mean()
    .sort_values()
    .index
)
```

```python
g = sns.factorplot('AveragePrice','region',data=df[filt],
                   hue='year',
                   size=8,
                   aspect=0.7,
                   palette='Blues',
                   order=order,
                   join=False,
              )
```

```python
# now let's look at more closely
# for now I take first 5 region
regions = cities[:5]
regions
```

```python
mask = (
    df['region'].isin(regions)
    & (df['type']=='conventional')
)
```

```python
df['Month'] = df['Date'].dt.month
df[mask].head()
```

```python
df['Year'] = df['Date'].dt.year
df[mask].head()
```

```python
g = sns.factorplot('Month','AveragePrice',data=df[mask],
               hue='Year',
               row='region',
               aspect=2,
               palette='Blues'
              )
```

```python
# now let's look at organic type
filt1 = df['type']=='organic'

order1 = (
    df[filt1 & (df['year']==2018)]
    .groupby('region')['AveragePrice']
    .mean()
    .sort_values()
    .index
)

g1 = sns.factorplot('AveragePrice','region',data=df[filt1],
                   hue='year',
                   size=8,
                   aspect=0.7,
                   palette='Blues',
                   order=order1,
                   join=False,
              )
```

```python
mask1 = (
    df['region'].isin(regions)
    & (df['type']=='organic')
)

g = sns.factorplot('Month','AveragePrice',data=df[mask1],
               hue='Year',
               row='region',
               aspect=2,
               palette='Blues'
              )
```
